```python
import pandas as pd
url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
df = pd.read_csv(url)
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
?plt.pie
```

```python
#plt.pie(df.sex.count)
# df.groupby(['sex']).count()
df1 = pd.Categorical(df['sex'])
df1.describe()
```

```python
# Pie Chart
plt.figure(figsize = (14,4))
plt.pie(df.sex.value_counts(),explode = [0.1,0],labels = ['Male','Female'],colors = ['g','gold'],
        autopct='%1.1f%%', shadow=True, startangle=45)
plt.axis('equal')
plt.title('Pie Chart')
plt.show()
```

```python
df.sex.value_counts()
```

```python
# Scatter Plot
plt.figure(figsize = (14,6))
gender = df.sex
gender_color = []
for i in gender:
    if i == 'male':
        gender_color.append('gold')
    else:
        gender_color.append('green')
plt.scatter(df.age,df.fare, c= gender_color)
plt.xlabel('Fair Paid')
plt.ylabel('Age')
plt.title('Scatter Plot')
plt.legend(gender)
plt.show()
```
